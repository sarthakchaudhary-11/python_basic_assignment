```python
#1Ô∏è‚É£ Reverse a string:
def reverse_string(s):
    return s[::-1]
```

```python
#2Ô∏è‚É£ Count the number of vowels in a string:
def count_vowels(s):
    vowels = "aeiouAEIOU"
    return sum(1 for char in s if char in vowels)
```

```python
# 3Ô∏è‚É£ Check if a given string is a palindrome:
def is_palindrome(s):
    return s == s[::-1]

```

```python
#4Ô∏è‚É£ Check if two strings are anagrams:
def are_anagrams(s1, s2):
    return sorted(s1) == sorted(s2)
```

```python
#5Ô∏è‚É£ Find all occurrences of a substring within another string:
def find_substring_occurrences(text, sub):
    return [i for i in range(len(text)) if text.startswith(sub, i)]
```

```python
#6Ô∏è‚É£ Perform basic string compression using counts of repeated characters:
def string_compression(s):
    compressed = []
    count = 1
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            compressed.append(s[i - 1] + str(count))
            count = 1
    compressed.append(s[-1] + str(count))
    return "".join(compressed) if len(compressed) < len(s) else s
```

```python
#7Ô∏è‚É£ Check if a string has all unique characters:
def has_unique_characters(s):
    return len(set(s)) == len(s)
```

```python
#8Ô∏è‚É£ Convert a given string to uppercase or lowercase:

def convert_case(s, to_upper=True):
    return s.upper() if to_upper else s.lower()
```

```python
# 9Ô∏è‚É£ Count the number of words in a string:

def count_words(s):
    return len(s.split())
```

```python
#üîü Concatenate two strings without using +:

def concatenate_strings(s1, s2):
    return "".join([s1, s2])
```

```python
def remove_element(lst, elem):
    # Write a code to remove all occurrences of a specific element from a list
    return [x for x in lst if x != elem]
```

```python
def second_largest(lst):
    # Implement a code to find the second largest number in a given list of integers
    unique_sorted = sorted(set(lst), reverse=True)
    return unique_sorted[1] if len(unique_sorted) > 1 else None
```

```python

def count_occurrences(lst):
    # Create a code to count the occurrences of each element in a list and return a dictionary with elements as keys and their counts as values
    return {x: lst.count(x) for x in set(lst)}
```

```python

def reverse_list_in_place(lst):
    # Write a code to reverse a list in-place without using any built-in reverse functions
    left, right = 0, len(lst) - 1
    while left < right:
        lst[left], lst[right] = lst[right], lst[left]
        left += 1
        right -= 1
```

```python

def remove_duplicates(lst):
    # Implement a code to find and remove duplicates from a list while preserving the original order of elements
    seen = set()
    return [x for x in lst if not (x in seen or seen.add(x))]
```

```python

def is_sorted(lst):
    # Create a code to check if a given list is sorted (either in ascending or descending order) or not
    return lst == sorted(lst) or lst == sorted(lst, reverse=True)
```

```python
def merge_sorted_lists(lst1, lst2):
    # Write a code to merge two sorted lists into a single sorted list
    result = []
    i, j = 0, 0
    while i < len(lst1) and j < len(lst2):
        if lst1[i] < lst2[j]:
            result.append(lst1[i])
            i += 1
        else:
            result.append(lst2[j])
            j += 1
    result.extend(lst1[i:])
    result.extend(lst2[j:])
    return result
```

```python
def list_intersection(lst1, lst2):
    # Implement a code to find the intersection of two given lists
    return list(set(lst1) & set(lst2))
```

```python
def list_union(lst1, lst2):
    # Create a code to find the union of two lists without duplicates
    return list(set(lst1) | set(lst2))
```

```python
def shuffle_list(lst):
    # Write a code to shuffle a given list randomly without using any built-in shuffle functions
    import random
    for i in range(len(lst) - 1, 0, -1):
        j = random.randint(0, i)
        lst[i], lst[j] = lst[j], lst[i]
```

```python

def common_tuple_elements(t1, t2):
    # Write a code that takes two tuples as input and returns a new tuple containing elements that are common to both input tuples
    return tuple(set(t1) & set(t2))
```

```python

def set_intersection():
    # Create a code that prompts the user to enter two sets of integers separated by commas and prints their intersection
    set1 = set(map(int, input("Enter first set of integers separated by commas: ").split(',')))
    set2 = set(map(int, input("Enter second set of integers separated by commas: ").split(',')))
    print("Intersection:", set1 & set2)
```

```python

def set_difference():
    # Develop a code that prompts the user to input two sets of strings and prints elements in the first set but not in the second
    set1 = set(input("Enter first set of strings separated by commas: ").split(','))
    set2 = set(input("Enter second set of strings separated by commas: ").split(','))
    print("Elements in first set but not in second:", set1 - set2)
```

```python
def slice_tuple(t, start, end):
    # Create a code that takes a tuple and two integers as input and returns a new tuple containing elements within the specified range
    return t[start:end]
```

```python

def set_union():
    # Write a code that prompts the user to input two sets of characters and prints their union
    set1 = set(input("Enter first set of characters separated by commas: ").split(','))
    set2 = set(input("Enter second set of characters separated by commas: ").split(','))
    print("Union:", set1 | set2)
```

```python

def merge_dictionaries(d1, d2):
    # Write a code that merges two dictionaries, adding values for common keys
    return {key: d1.get(key, 0) + d2.get(key, 0) for key in set(d1) | set(d2)}
```

```python
def access_nested_dict(d, keys):
    # Write a code to access a value in a nested dictionary
    for key in keys:
        if isinstance(d, dict) and key in d:
            d = d[key]
        else:
            return None
    return d
```

```python
def sort_dict_by_values(d, ascending=True):
    # Write a code that takes a dictionary and returns a sorted version based on values
    return dict(sorted(d.items(), key=lambda item: item[1], reverse=not ascending))
```

```python
def invert_dictionary(d):
    # Write a code that inverts a dictionary, swapping keys and values
    inverted = {}
    for key, value in d.items():
        inverted.setdefault(value, []).append(key)
    return inverted
```
